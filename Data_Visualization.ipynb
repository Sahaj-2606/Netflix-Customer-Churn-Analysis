```python
# Import and Load data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#pandas for data manipulation
#matplotlib for basic plots
#seaborn for advanced and aesthetic plots

# Display settings
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)

# Load dataset
path = r"C:\Users\udbha\.cache\kagglehub\datasets\abdulwadood11220\netflix-customer-churn-dataset\versions\1\netflix_customer_churn.csv"
df = pd.read_csv(path)
print(df.head())
```

```python
# Quick glance
df.info()
df.describe(include='all')
```

```python
# Univariate analysis
```

```python
#A. Numeric: Distribution & Histogram
sns.histplot(df['age'], kde=True, bins=20, color='skyblue')
plt.title("Age Distribution of Netflix Users")
plt.show()
```

```python
#B. Categorical count plot
sns.countplot(x='payment_method', data=df, palette='Set2')
plt.title("Gender Distribution")
plt.show()
```

```python
#C. Boxplot: Outlier detection
sns.boxplot(x='subscription_type', y='watch_hours', data=df, palette='cool')
plt.title("Monthly Fee by Subscription Type")
plt.show()
```

```python
#D. Violin plot (optional)
sns.violinplot(x='subscription_type', y='watch_hours', data=df, palette='muted')
plt.title("Distribution of Watch Hours by Subscription Type")
plt.show()
```

```python
# Bivariate Analysis
```

```python
#A. Scatter Plot (Continuous vs Continuous)
sns.scatterplot(x='age', y='watch_hours', hue='churned', data=df, alpha=0.7)
plt.title("Watch Hours vs Age (by Churn Status)")
plt.show()
```

```python
#B. Correlation Heatmap
corr = df[['age', 'watch_hours', 'last_login_days', 'monthly_fee', 
           'avg_watch_time_per_day', 'number_of_profiles', 'churned']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#C. Bar Plot (Categorical vs Numeric)
sns.barplot(x='region', y='churned', data=df, estimator=lambda x: sum(x)/len(x))
plt.title("Churn Rate by Region")
plt.ylabel("Churn Rate")
plt.xticks(rotation=45)
plt.show()
```

```python
# Multivariate Visualization
```

```python
#A. Pairplot (Relationships Across Features)
sns.pairplot(df[['age', 'watch_hours', 'monthly_fee', 'churned']], hue='churned')
plt.show()
```

```python
#B. FacetGrid (Multiple Small Plots)
g = sns.FacetGrid(df, col='region', hue='churned', height=3, col_wrap=3)
g.map(sns.scatterplot, 'watch_hours', 'avg_watch_time_per_day', alpha=0.7)
g.add_legend()
plt.show()
```

```python
#C. Heatmap by Category
pivot = df.pivot_table(values='churned', index='subscription_type', columns='region', aggfunc='mean')
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.title("Churn Rate by Subscription Type & Region")
plt.show()
```

```python
#Aesthetic Customization & Storytelling
plt.figure(figsize=(10,6))
sns.barplot(x='favorite_genre', y='watch_hours', hue='churned', data=df, palette='Spectral')
plt.title("Watch Hours by Genre and Churn Status")
plt.xticks(rotation=45)
plt.xlabel("Favorite Genre")
plt.ylabel("Average Watch Hours")
plt.legend(title="Churned", loc='upper right')
plt.tight_layout()
plt.show()
```

```python
#Bonus: Interactive Visualization with Plotly
import plotly.express as px

fig = px.scatter(df, x='watch_hours', y='monthly_fee',
                 color='subscription_type', size='avg_watch_time_per_day',
                 hover_data=['region', 'churned'],
                 title="Interactive Scatter: Watch Hours vs Monthly Fee")
fig.show()
```
