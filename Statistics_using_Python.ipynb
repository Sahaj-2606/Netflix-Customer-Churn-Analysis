```python
import pandas as pd
import numpy as np

data = {'Age': [23, 25, 31, 22, 35, 45, 52, 46, 21, 23, 34, 29]}
df1 = pd.DataFrame(data)

# Measures of center
mean_age = df1['Age'].mean()
median_age = df1['Age'].median()
mode_age = df1['Age'].mode()[0]

# Spread
std_age = df1['Age'].std()
var_age = df1['Age'].var()
range_age = df1['Age'].max() - df1['Age'].min()

# Percentiles
q1 = df1['Age'].quantile(0.25)
q3 = df1['Age'].quantile(0.75)
iqr = q3 - q1

print(f"Mean: {mean_age}, Median: {median_age}, Mode: {mode_age}")
print(f"Std: {std_age}, Variance: {var_age}, Range: {range_age}")
print(f"Q1: {q1}, Q3: {q3}, IQR: {iqr}")
```

```python
#Probability Essentials

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

# Binomial example: flipping a coin 10 times
n, p = 10, 0.5
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

plt.bar(x, pmf)
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Successes')
plt.ylabel('Probability')
plt.show()

# Normal distribution
x = np.linspace(-4, 4, 100)
plt.plot(x, norm.pdf(x, 0, 1))
plt.title('Normal Distribution (mean=0, std=1)')
plt.show()
```

```python
# Sampling & Confidence Interval
# Population simulation
population = np.random.normal(50, 10, 10000)

# Sampling
sample = np.random.choice(population, 100, replace=False)

# CLT demo
means = [np.mean(np.random.choice(population, 100)) for _ in range(1000)]

plt.hist(means, bins=30)
plt.title('Sampling Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()

# Confidence Interval (95%)
mean = np.mean(sample)

std_err = np.std(sample) / np.sqrt(len(sample))
from scipy.stats import norm
ci = norm.interval(0.95, loc=mean, scale=std_err)
print("95% CI:", ci)
```

```python
# Hypothesis Testing
from scipy.stats import ttest_ind, chi2_contingency

# t-test example
group_A = np.random.normal(60, 5, 30)
group_B = np.random.normal(63, 5, 30)
t_stat, p_val = ttest_ind(group_A, group_B)
print(f"T-statistic: {t_stat}, P-value: {p_val}")

# Chi-square example
data = [[30, 10], [20, 40]]  # e.g., Male/Female vs Purchase/No Purchase
chi2, p, dof, expected = chi2_contingency(data)
print(f"Chi-square: {chi2}, p-value: {p}")
```

```python
# Correlation vs. Causation
np.random.seed(42)
x = np.arange(50)
y = x * 2 + np.random.normal(0, 5, 50)
z = np.random.normal(0, 50, 50)

df_corr = pd.DataFrame({'x': x, 'y': y, 'z': z})
df_corr
print(df_corr.corr())

plt.scatter(df_corr['x'], df_corr['y'])
plt.title('Correlation between X and Y')
plt.show()
```

```python
# # Regression I (Simple Linear Regression)

import statsmodels.api as sm

X = df_corr['x']
y = df_corr['y']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())

plt.scatter(df_corr['x'], df_corr['y'])
plt.plot(df_corr['x'], model.predict(X), color='red')
plt.title('Simple Linear Regression')
plt.show()
```

```python
# Regression II (Multiple Regression)
np.random.seed(42)
X1 = np.random.rand(50)
X2 = np.random.rand(50)
y = 3 + 2*X1 + 4*X2 + np.random.normal(0, 0.3, 50)
df_multi = pd.DataFrame({'X1': X1, 'X2': X2, 'y': y})

X = sm.add_constant(df_multi[['X1', 'X2']])
model = sm.OLS(df_multi['y'], X).fit()
print(model.summary())
```
